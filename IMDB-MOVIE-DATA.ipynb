## Classify Movies Based on Rating [Excellent , Good ,Average]

```python
df.columns
```

```python
def rating(rating):
    if rating >=7.0:
        return "Excellent"
    elif rating >=6.0:
        return "Good"
    else:
        return "Average"
```

```python
df["Rating_cat"]=df['Rating'].apply(rating)
```
